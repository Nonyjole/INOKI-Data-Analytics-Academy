```python
# 2: WHAT IS THE AVERAGE NUMBER OF GAMES WON BY EACH TAEM

# Calculate the number of games won by each team in both "team1" and "team2" columns
games_won_team1 =  T20_Dataset[ T20_Dataset['winner'] ==  T20_Dataset['team1']]['team1'].value_counts()
games_won_team2 =  T20_Dataset[ T20_Dataset['winner'] ==  T20_Dataset['team2']]['team2'].value_counts()

# Combine the counts from both columns and calculate the mean
total_games_won = games_won_team1.add(games_won_team2, fill_value=0)
mean_games_won = total_games_won.mean()

print("The mean number of games won by each team:", mean_games_won, 'wins')
```

```python
# 3: WHAT IS THE AVERAGE NUMBER OF GAMES PLAYED BY EACH TAEM
# Calculate the number of games played by each team in both "team1" and "team2" columns
games_played_team1 = T20_Dataset['team1'].value_counts()
games_played_team2 = T20_Dataset['team2'].value_counts()

# Combine the counts from both columns and calculate the mean
total_games_played = games_played_team1.add(games_played_team2, fill_value=0)
mean_games_played = total_games_played.mean()

print("The mean number of games played by each team is:", mean_games_played, 'games')
```

```python
# 4: WHAT IS THE NUMBER OF MATCHES PLAYED BY EACH TEAM

Number_Of_matches_Played_By_each_Team = T20_Dataset['team1'].value_counts() + T20_Dataset['team2'].value_counts()
Number_Of_matches_Played_By_each_Team
```

```python
# Use pie chart to show the percentage of matches played by each team

colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'pink', 'grey', 'lime']

# Plotting the data as a pie chart
plt.figure(figsize=(5, 4))
plt.pie(Number_Of_matches_Played_By_each_Team, labels=None, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('THE PERCENTAGE OF MATCHES PLAYED BY EACH TEAM')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.

# Move the legend away from the pie chart
plt.legend(Number_Of_matches_Played_By_Each_Team .index, loc='center left', bbox_to_anchor=(2, 0.5))

plt.tight_layout()
plt.show()
```

```python
# 7: FIND EACH TEAM AND THEIR TOTAL NUMBER OF GAMES WON BY RUNS AND WICKETS



# Combine team1 and team2 columns to get all unique team names
unique_teams = pd.concat([T20_Dataset['team1'], T20_Dataset['team2']]).unique()

# Create a DataFrame to store team data
team_data = pd.DataFrame({'Team': unique_teams, 'Total_Wins_By_Runs': 0, 'Total_Wins_By_Wickets': 0})

# Loop through the rows in the dataset and update win counts for each team
for index, row in T20_Dataset.iterrows():
    if row['winner'] == row['team1']:
        team_data.loc[team_data['Team'] == row['team1'], 'Total_Wins_By_Runs'] += 1
    elif row['winner'] == row['team2']:
        team_data.loc[team_data['Team'] == row['team2'], 'Total_Wins_By_Wickets'] += 1

print(team_data)
```

```python
# Visualize the distribution of total wins by runs and wickets for each unique team using a grouped bar chart

# Plotting the data as a grouped bar chart
plt.figure(figsize=(12, 6))
bar_width = 0.4
bar_positions = range(len(team_data))

plt.bar(bar_positions, team_data['Total_Wins_By_Runs'], width=bar_width, label='Wins by Runs', color='c')
plt.bar([pos + bar_width for pos in bar_positions], team_data['Total_Wins_By_Wickets'], width=bar_width, label='Wins by Wickets', color='m')

plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Total Number of Wins by Runs and Wickets for Each Team')
plt.xticks([pos + bar_width/2 for pos in bar_positions], team_data['Team'], rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# 8: WHICH PLAYER(S) HAD THE HIGHEST PLAYER OF THE MATCH


# # Calculate the count of Player of the Match awards for each player
player_of_match_counts = T20_Dataset['player of the match'].value_counts()

# Find the player(s) with the highest count of Player of the Match awards
highest_player_of_match_count = player_of_match_counts.max()
players_with_highest_count = player_of_match_counts[player_of_match_counts == highest_player_of_match_count].index.tolist()

print("Player(s) with the highest Player of the Match awards:", players_with_highest_count)
```
